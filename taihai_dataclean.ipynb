```python
import polars as pl

# 讀取 CSV 檔案
df = pl.read_csv(r"C:\Users\Leon\Desktop\taihai_news_updated.csv")

# 將 Polars DataFrame 轉換為 Pandas DataFrame
pandas_df = df.to_pandas()

# 顯示所有資料
print(pandas_df)
```

```python
# 計算 'content' 欄位中含有 '�' 的行數
rows_with_issues = df['content'].str.contains('�').sum()

# 顯示結果
print(f"'content' 欄位中含有 '�' 的行數: {rows_with_issues}")
```

```python
import pandas as pd

# 假設你已經加載了 DataFrame df
# 刪除 'content' 欄位中包含 '�' 的行
df = df[df['content'].apply(lambda x: '�' not in str(x))]

# 顯示結果
display(df.head())
```

```python
# 計算 'content' 欄位中含有 '�' 的行數
rows_with_issues = df['content'].str.contains('�').sum()

# 顯示結果
print(f"'content' 欄位中含有 '�' 的行數: {rows_with_issues}")
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

# 假設你已經加載了 DataFrame df
# 找出 'content' 為 None 的行
urls_to_revisit = df[df['content'].isna()]['url']

# 重新爬取這些 URLs
for url in urls_to_revisit:
    try:
        response = requests.get(url)
        if response.status_code == 200:
            response.encoding = 'utf-8'  # 設定編碼格式
            html_content = response.text
            soup = BeautifulSoup(html_content, 'html.parser')
            
            # 假設內容在 'rm_txt_con' class 的 div 內
            article = soup.find('div', class_='rm_txt_con')
            
            # 如果找到內容，提取純文本並更新 'content'
            if article:
                text_content = article.get_text()
                df.loc[df['url'] == url, 'content'] = text_content
                print(f"已更新 {url} 的 content")
            else:
                print(f"未能找到正文內容: {url}")
        else:
            print(f"無法抓取網頁，狀態碼: {response.status_code}，URL: {url}")
    except Exception as e:
        print(f"處理 URL {url} 時發生錯誤: {e}")

# 顯示更新後的資料
display(df.head())
```

```python
# 計算 'content' 欄位為 None 的行數
none_content_count = df['content'].isna().sum()

# 顯示結果
print(f"content 欄位為 None 的數量：{none_content_count}")
```
